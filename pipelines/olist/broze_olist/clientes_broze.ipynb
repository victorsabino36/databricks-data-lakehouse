```python
from pyspark.sql.functions import current_timestamp
```

```python
df = spark.read.csv("/Volumes/data/lainding/lainding/dataset-olist/olist_customers_dataset.csv", header=True)
```

```python
df.display()
```

```python
df_date_ingestao = df.withColumn("data_ingestao", current_timestamp())

df_date_ingestao.printSchema()
```

```python
%sql
CREATE TABLE IF NOT EXISTS workspace.olist_bronze.clientes_bronze (
 customer_id string, 
 customer_unique_id string, 
 customer_zip_code_prefix string,
 customer_city string, 
 customer_state string, 
 data_ingestao timestamp 
)
```

```python
df_date_ingestao.write.mode("overwrite").saveAsTable("workspace.olist_bronze.clientes_bronze")
```

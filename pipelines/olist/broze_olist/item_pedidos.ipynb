```python
from pyspark.sql.functions import current_timestamp
```

```python
df = spark.read.csv("/Volumes/data/lainding/lainding/dataset-olist/olist_order_items_dataset.csv", header=True)
```

```python
df.display()
```

```python
df_data_ingestao = df.withColumn("data_ingestao", current_timestamp())
```

```python
df_data_ingestao.printSchema()
```

```python
%sql
CREATE TABLE IF NOT EXISTS workspace.olist_bronze.itens_pedido_broze(
  order_id string,
  order_item_id string,
  product_id string,
  seller_id string,
  shipping_limit_date string,
  price string,
  freight_value string,
  data_ingestao timestamp
)
```

```python
df_data_ingestao.write.mode("overwrite").saveAsTable("workspace.olist_bronze.itens_pedido_broze")


```

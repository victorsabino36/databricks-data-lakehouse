```sql
CREATE OR REPLACE MATERIALIZED VIEW workspace.olist_bronze.view_pedidos as
select 
  ped.order_id id_pedido
, ped.customer_id id_cliente
, cli.customer_state uf_cliente
, cli.customer_city cidade_cliente
, ped.order_status status_pedido
, TO_DATE(ped.order_purchase_timestamp) AS dt_pedido
, DATE_FORMAT(ped.order_purchase_timestamp, 'HH:mm:ss') AS hr_pedido
, CASE WHEN ped.order_approved_at is null then 'negado' else 'aprovado' end as status_aprovacao
, date(ped.order_delivered_carrier_date) data
, ped.order_delivered_customer_date data_entrega
, date_diff(DAY, ped.order_approved_at, order_delivered_customer_date) dias_entrega_efutuada
, date(ped.order_estimated_delivery_date) data_prevista_entrega
, date_diff(DAY,ped.order_approved_at, ped.order_estimated_delivery_date  ) prazo_entrega
, pag.payment_value valor_pagamento
, pag.payment_type tipo_pagamento
from `workspace`.`olist_bronze`.pedidos_bronze ped
left join `workspace`.`olist_bronze`.pagamentos_bronze pag on ped.order_id = pag.order_id
left join `workspace`.`olist_bronze`.clientes_bronze cli on ped.customer_id = cli.customer_id
--left join `workspace`.`olist_bronze`.itens_pedido_broze item on ped.order_id = item.order_id
order by id_pedido desc
```
